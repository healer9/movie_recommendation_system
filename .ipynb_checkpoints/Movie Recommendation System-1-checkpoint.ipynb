# Creating Movie Recommendation System

```python
ratings.head()
```

```python
movie_matrix = df.pivot_table(index='user_id', columns='title', values='rating')
```

```python
movie_matrix.head(n=10)
```

```python
ratings.sort_values('num of ratings', ascending=False).head()
```

```python
starwars_user_ratings = movie_matrix['Star Wars (1977)']
```

```python
starwars_user_ratings.head()
```

```python
# Cor relationg of all movies with start wars movie

similar_to_starwars = movie_matrix.corrwith(starwars_user_ratings)
```

```python
similar_to_starwars
```

```python
corr_of_starwars = pd.DataFrame(similar_to_starwars, columns=["Correlations"])
```

```python
corr_of_starwars.dropna(inplace=True)
```

```python
corr_of_starwars
```

```python
corr_of_starwars.sort_values('Correlations',ascending=False).head(n=10)
```

```python
# filtration of movies based upon how many users have watch it

corr_of_starwars = corr_of_starwars.join(ratings['num of ratings'])
corr_of_starwars.head()
```

```python
corr_of_starwars[corr_of_starwars['num of ratings'] > 100].sort_values('Correlations', ascending=False)
```

# Predict Function

```python
def predict_movies(movie_name):
    movie_user_ratings = movie_matrix[movie_name]
    similar_to_movie = movie_matrix.corrwith(movie_user_ratings)
    
    corr_of_movie = pd.DataFrame(similar_to_movie, columns=["Correlations"])
    corr_of_movie.dropna(inplace=True)
    
    corr_of_movie = corr_of_movie.join(ratings['num of ratings'])
    
    predictions = corr_of_movie[corr_of_movie['num of ratings'] > 100].sort_values('Correlations', ascending=False)
    
    return predictions
```

```python
predictions = predict_movies("Titanic (1997)")
```
